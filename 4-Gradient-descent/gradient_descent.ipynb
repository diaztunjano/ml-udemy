# Gradient Descent

#### 1. Importing libraries


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### 2. Simple cost function

$$f(x) = x^2 + x +1$$
$$df(x) = 2x + 1$$

```python
def f(x):
    return x**2 + x + 1

x_1 = np.linspace(start=-3, stop=3, num=100)
```

```python
## Plot
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

```python
## Derivative
def df(x):
    return 2*x + 1

## Plot function and derivative side by side
plt.figure(figsize=[15, 5])

# Subplot 1 - Cost function
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st subplot
plt.title('Cost function', fontsize=17)
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))

# Subplot 2 - Derivative
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd subplot
plt.title('Slope of the cost function', fontsize=17)
plt.xlim([-2, 3])
plt.ylim(-3, 6)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.plot(x_1, df(x_1), color='red', linewidth=3)

plt.show()


```

#### 3. Gradient Descent

```python
# Gradient Descent Algorithm with Loop:
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.0001

# Lists:
x_list = [new_x]
slope_list = [df(new_x)]

for n in range(50):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient

    x_list.append(new_x)
    slope_list.append(df(new_x))

    if abs(new_x - previous_x) < precision:
        print(f"Loop ran this many times: {n}")
        break

    if n % 5 == 0:
        print(f"Previous x: {previous_x}, new x: {new_x}, gradient: {gradient}")

print(f"The local minimum occurs at {new_x}")
print(f"The slope or df(x) value at this point is {df(new_x)}")
```

```python
## Superimpose the gradient descent calculations on plot
plt.figure(figsize=[20, 5])


# Subplot 1 - Cost function
plt.subplot(1, 3, 1) # 1 row, 2 columns, 1st subplot
plt.title('Cost function', fontsize=17)
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))

plt.scatter(x_list, f(np.array(x_list)), color='red', s=100, alpha=0.6)

# Subplot 2 - Derivative
plt.subplot(1, 3, 2) # 1 row, 2 columns, 2nd subplot
plt.title('Slope of the cost function', fontsize=17)
plt.xlim([-2, 3])
plt.ylim(-3, 6)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.plot(x_1, df(x_1), color='red', linewidth=3)

plt.scatter(x_list, slope_list, color='green', s=100, alpha=0.6)

# Subplot 3 - Derivative (Close up)
plt.subplot(1, 3, 3) # 1 row, 2 columns, 2nd subplot
plt.title('Slope of the cost function', fontsize=17)
plt.xlim([-0.55, -0.2])
plt.ylim(-0.3, 0.8)
plt.grid()
plt.plot(x_1, df(x_1), color='red', linewidth=3)
plt.scatter(x_list, slope_list, color='green', s=100, alpha=0.6)



plt.show()
```
